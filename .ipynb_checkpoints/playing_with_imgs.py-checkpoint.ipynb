```python
import cv2

import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Getting data from matlab file
matlab_data = sio.loadmat('/Users/abhimanyusuthar/nyu/studentproj2/code/data/studentdata1.mat')
DATA = matlab_data['data']

```

```python
img = DATA['img'][0][0]
img2 = DATA['img'][0][50]
plt.imshow(img)
```

```python
combined_image = cv2.hconcat([img,img2])
plt.axis("off")
plt.imshow(combined_image)
```

```python
alpha =0.4
beta = (1.0 - alpha)
dst = cv2.addWeighted(img, alpha, img2, beta, 0.0)
dst = np.uint8(alpha*(img)+beta*(img2))
plt.imshow(dst)
```
